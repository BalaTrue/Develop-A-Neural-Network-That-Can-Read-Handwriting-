**Importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

**loading the dataset**

```python
objects= tf.keras.datasets.mnist
```

**Dividing the images and their outputs into training and testing sets**

```python
(training_images, training_outputs), (testing_images, testing_outputs) = objects.load_data()
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

**Cheking the data set images, which needed to be identified by the neural network**

```python
for i in range (9):
    plt.subplot(330 + 1 + i)
    plt.imshow(training_images[i])
```

**Printing the data images into metrics form**

```python
print(training_images.shape)
print(training_images[0])
```

**Normalizing the matrix array of  number images**

```python
training_images = training_images/255.0
test_images = testing_images/255.0
```

download.png

**Flattening the 2-dimensional array into one dimensional array or a single column; which will behave as 1st inpur layer for neural network**

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(input_shape=(28,28)),
                                    tf.keras.layers.Dense(128, activation ='relu'),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

**Dense function used to make sure successful conection of each neurons with each other**

maxresdefault.jpg

**Using of AdomOptimizer**

```python
model.compile(optimizer = tf.keras.optimizers.Adam(),
              loss ='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

**Fitting of the training data into the model for 5 iterations**

```python
model.fit(training_images,training_outputs, epochs=5)
```

**98% Accuracy achieved with 5 iterations**

```python
model.fit(training_images,training_outputs, epochs=9)
```

**99% Accuracy achieved with 9 iterations**

```python
print(model.evaluate(testing_images,testing_outputs))
```

**97 % of total accuracy of our neural network has been achieved**

**Its time to test the neural network**

```python
plt.imshow(testing_images[0])
```

**Test image**

```python
prediction=model.predict(testing_images)
print(np.argmax(prediction[0]))
```

**Our model predicted it as a 7 number, which is accurate**
